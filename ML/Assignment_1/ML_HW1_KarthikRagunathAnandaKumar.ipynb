```python
# Utility Function To Check Convergence
```

```python
def check_convergence(W, X, b, Y, M):
  check_classifications = np.multiply(Y, np.add(np.sum(np.multiply(W, X), axis = 1).reshape(-1, 1), b)).reshape(M,)
  valid = 0
  for i in range(M):
    if check_classifications[i] > 0:
      valid += 1
  if valid == M:
    return True
  else:
    return False
```

```python
X = perceptron_df[['x1', 'x2', 'x3', 'x4']].to_numpy()
Y = perceptron_df[['y']].to_numpy()
Y = Y.reshape(-1,1)
step_size = 1
num_epochs = 0
M = len(perceptron_df)
W = np.array([0.0, 0.0, 0.0, 0.0])
b = 0
while True:
  # loss = (max{0, (1 - y(wx + b))})^2
  WX = np.sum(np.multiply(W, X), axis=1) # 999 * 1
  linear_sum = WX + b # 999 * 1
  y_linear_sum_mult = np.multiply(Y, linear_sum.reshape(-1, 1)) # 999 * 1
  y_linear_sum_mult = (1 - y_linear_sum_mult) # 999 * 1
  perceptron_loss = np.maximum(y_linear_sum_mult, 0) ** 2 ### 999 * 1
  perceptron_loss = (1 / M) * np.sum(perceptron_loss) # scalar
  pre_compute = np.multiply(2, np.maximum(0, y_linear_sum_mult))
  w1_grad = np.sum(np.multiply(pre_compute, np.multiply(-1, np.multiply(Y, X[:, 0].reshape(-1, 1))))) / M 
  w2_grad = np.sum(np.multiply(pre_compute, np.multiply(-1, np.multiply(Y, X[:, 1].reshape(-1, 1))))) / M
  w3_grad = np.sum(np.multiply(pre_compute, np.multiply(-1, np.multiply(Y, X[:, 2].reshape(-1, 1))))) / M 
  w4_grad = np.sum(np.multiply(pre_compute, np.multiply(-1, np.multiply(Y, X[:, 3].reshape(-1, 1))))) / M 
  b_grad = np.sum(np.multiply(pre_compute, np.multiply(-1, Y))) / M
  W = np.subtract(W, np.multiply(step_size, np.array([w1_grad, w2_grad, w3_grad, w4_grad])))
  b = float(b) - (step_size * b_grad)
  print(W, b)
  num_epochs += 1
  if check_convergence(W, X, b, Y, M):
    print("Converges at:", num_epochs)
    break
  if num_epochs == 3:
    break
```

```python
M = 2
X = np.array([[1, 1, 1], [1, 1, -1]]).reshape(2, 3)
Y = np.array([1,-1]).reshape(2, 1)
step_size = 0.01
num_epochs = 0
W = np.array([0.0, 0.0, 0.0])
b = 0
while True:
  WX = np.sum(np.multiply(W, X), axis=1)
  linear_sum = WX + b
  y_linear_sum_mult = np.multiply(Y, linear_sum.reshape(-1, 1)) * -1
  perceptron_loss = np.maximum(y_linear_sum_mult, 0)
  perceptron_loss = (1 / M) * np.sum(perceptron_loss)
  w1_grad = 0
  w2_grad = 0
  w3_grad = 0
  b_grad = 0
  for index in range(M):
    if y_linear_sum_mult[index] >= 0:
      w1_grad += (-1 * Y[index] * X[index, 0])
      w2_grad += (-1 * Y[index] * X[index, 1])
      w3_grad += (-1 * Y[index] * X[index, 2])
      b_grad += (-1 * Y[index])
  W[0] = float(W[0]) - (float(step_size) * float(w1_grad))
  W[1] = float(W[1]) - (float(step_size) * float(w2_grad))
  W[2] = float(W[2]) - (float(step_size) * float(w3_grad))
  b = float(b) - (step_size * b_grad)
  num_epochs += 1
  if check_convergence(W, X, b, Y, M):
    print(W, b)
    print("Converges at:", num_epochs)
    break
```

```python
## Optimal Weight Computation
```

```python
mystery_df = pd.read_csv('mystery.data', names=["x1", "x2", "x3", "x4", "y"], header=None)
mystery_df_positive = mystery_df[mystery_df["y"] == 1]
mystery_df_negative = mystery_df[mystery_df["y"] == -1]
X = mystery_df[["x1", "x2", "x3", "x4"]].to_numpy()
Y = mystery_df[["y"]].to_numpy()
n = X.shape[0]
H = np.multiply(np.matmul(Y, Y.T), np.matmul(X, X.T)**3)
q = np.repeat([-1.0], n)
A = Y.reshape(1, -1)
b = 0.0
G = np.negative(np.eye(n))
h = np.zeros(n)
A = A.astype('float')
P = matrix(H)
q = matrix(q)
G = matrix(G)
h = matrix(h)
A = matrix(A)
b = matrix(b)
sol = solvers.qp(P, q, G, h, A, b)
alphas = np.array(sol["x"])
optimal_w = np.dot((Y * alphas).T, X)[0]
```

```python
optimal_w
```

```python
## Optimal bias calculation
```

```python
X_pos = mystery_df_positive[["x1", "x2", "x3", "x4"]].to_numpy()
linear_postive = np.sum(np.multiply(optimal_w, X_pos), axis=1)
X_neg = mystery_df_negative[["x1", "x2", "x3", "x4"]].to_numpy()
linear_negative = np.sum(np.multiply(optimal_w, X_neg), axis=1)

max_index_neg = np.argmax(linear_negative)
min_index_pos = np.argmin(linear_postive)
bias = (linear_postive[min_index_pos] + linear_negative[max_index_neg]) * -1 / 2
```

```python
print("Optimal Weight:", optimal_w, "Optimal Bias:", bias)
```

```python
## Optimal Margin Computation
```

```python
w_norm = LA.norm(optimal_w)
optimal_margin = 2 / w_norm
print("Optimal Margin:", optimal_margin)
```
